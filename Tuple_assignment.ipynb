Assignment 15: Tuple Functions

Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

```python
def get_tuple_stats(tpl):
    mini = min(tpl)
    maxi = max(tpl)
    total_sum = sum(tpl)
    return mini, maxi, total_sum

my_tpl = (10, 20, 30, 40, 50)
mi, ma, sm = get_tuple_stats(my_tpl)
print(f"Tuple: {my_tpl}")
print(f"Minimum: {mi}")
print(f"Maximum: {ma}")
print(f"Sum: {sm}")
```
